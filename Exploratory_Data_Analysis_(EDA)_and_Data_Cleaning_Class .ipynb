**Scatter Plot**


*   Helps visualize relationship between two variables. (The columns are the variables in this dataset)

*   Important visualization before running a regression model.
*   Helpful in other models as well



**Scatterplot Examples**

image.png



*   Positive Relationship: As the value of the X (horizontal) axis increases, the value of the Y (vertical) axis increases.
*   Negative Relationship: As the value of the X (horizontal) axis increases, the value of the Y (vertical) axis decreases.
*   No Relationship: No pattern.

**Box Plot**


*   Also known as whisker plot.
*   Gives us the distribution of a variable. (Columns of this dataset)
*   Points out the median, quantiles and outliers of the dataset.



**Interpretation of a Box Plot**

image.png



*   Minimum: It is the minimum value in the dataset excluding the outliers.
*   First Quartile (Q1): 25% of the data lies below the First (lower) Quartile.
*   Median (Q2): It is the mid-point of the dataset. Half of the values lie below it and half above.
*   Third Quartile (Q3): 75% of the data lies below the Third (Upper) Quartile.
*   Maximum: It is the maximum value in the dataset excluding the outliers.



```python
sns.boxplot(data=cars,x='Price')
```

```python
sns.boxplot(data=cars,x='Price',showfliers=False)             #returns boxplot without the outliers
```

**Necessary Measures for Outlier Detection**


*   Inter Quartile Range (IQR): Q3-Q1
*   Lower Limit = Q1 - 1.5*IQR
*   Upper Limit = Q3 + 1.5*IQR



```python
Q1=cars['Price'].quantile(.25)
Q3=cars['Price'].quantile(.75)
IQR = Q3 - Q1
print(Q1, Q3, IQR)
```

**Outlier Detection**



*   Values less than the lower limit are outliers.
*   Values more than the Upper limit are outliers.



```python
lowerLimit=Q1-1.5*IQR
upperLimit=Q3+1.5*IQR
print(lowerLimit,upperLimit)
```

**Outliers**

```python
cars[(cars['Price']<lowerLimit) | (cars['Price']>upperLimit)]
```

**Remove Outliers**

```python
priceOutlierIndex = cars[(cars['Price']<lowerLimit) | (cars['Price']>upperLimit)].index
cars.drop(priceOutlierIndex,inplace=True)
cars.nlargest(5,columns='Price')
```

**Same Steps Repeated for HP**

```python
sns.boxplot(data=cars,x='HP')
```

```python
sns.boxplot(data=cars,x='HP',showfliers=False)
```

```python
Q1=cars['HP'].quantile(.25)
Q3=cars['HP'].quantile(.75)
IQR = Q3 - Q1
lowerLimit=Q1-1.5*IQR
upperLimit=Q3+1.5*IQR
HPOutlierIndex = cars[(cars['HP']<lowerLimit) | (cars['HP']>upperLimit)].index
cars.drop(HPOutlierIndex,inplace=True)
cars.nlargest(5,columns='HP')
```
