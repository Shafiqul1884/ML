```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/data maining lab/heart.csv')
df
```

```python
df.corr()
```

```python
df.isnull().sum()
```

```python
x = df.drop('target',axis=1)
x
```

```python
y=df['target']
y
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
fit_features = SelectKBest(score_func = f_classif)
fit_features.fit(x,y)
```

```python
fs = pd.DataFrame(fit_features.scores_,index=x.columns,columns=['Score value'])

fs
```

```python
fs=fs.nlargest(9,'Score value')
fs
```

```python
from sklearn.ensemble import ExtraTreesClassifier
model=ExtraTreesClassifier()
model.fit(x,y)
```

```python
fs=pd.DataFrame(model.feature_importances_,index=x.columns,columns=['score values'])
fs
```

```python
fs=fs.nlargest(9,'score values')
fs
```

```python
df=df.join(fs)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sb

plt.figure(1, figsize=(10, 5))
sb.heatmap(df.corr(), annot=True, cmap='viridis')
plt.title('Heatmap for the Data')
plt.show()
```
