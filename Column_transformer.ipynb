```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
df=pd.read_csv('/content/drive/MyDrive/Dataset/covid_toy.csv')
df.head(10)
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns=['has_covid']),df.has_covid,test_size=0.2,random_state=42)
```

```python
ct = ColumnTransformer(transformers=[('ohe',OneHotEncoder(drop='first',sparse_output=False),['gender','city']),
                                     ('oe',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
                                     ('si',SimpleImputer(),['fever']),
                                    ],remainder='passthrough')
```

```python
ct.fit_transform(x_train)
```

```python
ct.transform(x_test)
```

```python
le=LabelEncoder()
y_train=le.fit_transform(y_train)
y_test=le.transform(y_test)
```
