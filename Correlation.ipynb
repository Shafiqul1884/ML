```python
import pandas as pd

# Load the dataset
file_path = '/content/Copy of AirQualityUCI.csv'
data = pd.read_csv(file_path)

# Select numeric columns and calculate correlations
numeric_data = data.select_dtypes(include=['float64', 'int64']).dropna(axis=1, how='all')
correlations = numeric_data.corr()

# Extract correlations with the target column "CO(GT)"
target_correlations = correlations["CO(GT)"].sort_values(ascending=False)

# Display the correlations
print(target_correlations)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a heatmap for correlations with the target "CO(GT)"
plt.figure(figsize=(8, 6))
sns.heatmap(
    target_correlations.to_frame(),
    annot=True,
    cmap="coolwarm",
    cbar=True,
    fmt=".2f",
    linewidths=0.5
)
plt.title("Correlation Heatmap with CO(GT)", fontsize=14)
plt.xlabel("CO(GT)", fontsize=12)
plt.ylabel("Features", fontsize=12)
plt.tight_layout()
plt.show()
```
