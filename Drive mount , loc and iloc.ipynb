# Question 2: Find out versicolor category with all attribute values and print it, then  print only null values of versicolor category.
# Solution:

```python
df1 = pd.read_csv("/content/drive/MyDrive/Dataset/Dataset-20240827T051357Z-001/ iris_verciclolor_with_nullvalues.csv")
```

```python
df1.isnull().sum()
```

```python
versicolor_data = df1[df1['species'] == 'versicolor']
print(versicolor_data)
```

```python
null_versicolor_data = versicolor_data[versicolor_data.isnull().any(axis=1)]
print(null_versicolor_data)
```
