```python
correlations = data[numerical_columns].corr()

# Extract correlations with the target column "CO(GT)"
target_correlations = correlations["Production (GWh)"].sort_values(ascending=False)

# Display the correlations
print(target_correlations)
```

```python
# Create a heatmap for correlations with the target "CO(GT)"
plt.figure(figsize=(8, 10))
sns.heatmap(
    target_correlations.to_frame(),
    annot=True,
    cmap="coolwarm",
    cbar=True,
    fmt=".2f",
    linewidths=0.5
)
plt.title("Correlation Heatmap with CO(GT)", fontsize=14)
plt.xlabel("CO(GT)", fontsize=12)
plt.ylabel("Features", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestRegressor
```

```python
# Step 2: Encode categorical variables
categorical_columns = data.select_dtypes(include=['object']).columns
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col].astype(str))
    label_encoders[col] = le
```

```python
target_column = 'Production (GWh)'
X = data.drop(columns=[target_column])
y = data[target_column]
```

```python
# Step 3: Normalize/scale numeric variables
scaler = StandardScaler()
numerical_columns = X.select_dtypes(include=['float64', 'int64']).columns
data[numerical_columns] = scaler.fit_transform(data[numerical_columns])
```

```python

# Using Recursive Feature Elimination (RFE) with Random Forest Regressor
rfe_model = RandomForestRegressor(random_state=42, n_estimators=100)
rfe = RFE(estimator=rfe_model, n_features_to_select=30)
X_selected = rfe.fit_transform(X, y)

# Get selected feature names
selected_features = X.columns[rfe.support_]
print("Top 30 Selected Features:")
print(selected_features)
```

```python
sdf = pd.DataFrame(X, columns=selected_features)
sdf = pd.concat([sdf, y], axis=1)
sdf
```

```python
from sklearn.model_selection import train_test_split
target_column = 'Production (GWh)'
a = sdf.drop(columns=[target_column])
b = sdf[target_column]
# Split the data into training (70%) and testing (30%)
X_train, X_test, y_train, y_test = train_test_split(a,b, test_size=0.3, random_state=42)
```
