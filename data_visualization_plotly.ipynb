```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load the Titanic dataset (Kaggle's sample Titanic dataset)
titanic_data_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(titanic_data_url)



# Task 1: Passenger Survival Analysis (Bar Chart with Percentages)
survival_counts = titanic['Survived'].value_counts()
# Bar chart
plt.bar(['Not Survived', 'Survived'], survival_counts, color=['red', 'green'])

# Annotate percentages
total = survival_counts.sum()
for i, count in enumerate(survival_counts):
    plt.text(i, count + 5, f'{(count / total * 100):.1f}%', ha='center')

# Add labels
plt.title('Passenger Survival Analysis')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
import seaborn as sns

# Group data by Pclass and Survival
pclass_survival = titanic.groupby(['Pclass', 'Survived']).size().unstack()

# Stacked bar chart
pclass_survival.plot(kind='bar', stacked=True, color=['red', 'green'])
plt.title('Class-wise Survival Rates')
plt.ylabel('Number of Passengers')
plt.xlabel('Passenger Class')
plt.legend(['Not Survived', 'Survived'])
plt.show()
```

```python
import plotly.express as px

# Create a sunburst chart
fig = px.sunburst(titanic, path=['Sex', 'Pclass', 'Survived'],
                  color='Survived', color_continuous_scale=['red', 'green'],
                  title='Gender-wise Survival Comparison')
fig.show()
```

```python
plt.hist(titanic[titanic['Survived'] == 1]['Age'].dropna(), bins=20, alpha=0.5, label='Survived', color='green')
plt.hist(titanic[titanic['Survived'] == 0]['Age'].dropna(), bins=20, alpha=0.5, label='Not Survived', color='red')

# Add labels and legend
plt.title('Age Distribution: Survivors vs Non-Survivors')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Fare', data=titanic)
plt.title('Fare Distribution Across Classes')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```
