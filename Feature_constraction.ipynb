# Applying Feature Construction

```python
x['Family_size']=x['SibSp']+x['Parch']+1
```

```python
x.head()
```

```python
def myfunc(num):
  if num==1:
    return 0
  elif num>1 and num<=4:
    return 1
  else:
    return 2
```

```python
x['Family_type']=x['Family_size'].apply(myfunc)
```

```python
x.head()
```

```python
x.drop(columns=['SibSp','Parch','Family_size'],inplace=True)
```

```python
x.head()
```

```python
np.mean(cross_val_score(LogisticRegression(), x, y, cv=200, scoring='accuracy'))
```
